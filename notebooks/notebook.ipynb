# Visualização Parcial dos Dados

Esta seção mostra:
- Uma amostra (head) dos dados
- Contagem de valores ausentes por coluna
- Distribuições básicas para 1 coluna numérica e 1 categórica (quando existirem)


```python
# Visualização parcial do CSV
import pandas as pd
import plotly.express as px
import os

# Caminho do arquivo (relativo à pasta do notebook)
csv_path = os.path.join('carf_julgamentos_2024.csv')

# Carregar dados com dtype flexível para evitar erros por colunas mistas
try:
    df = pd.read_csv(csv_path, low_memory=False)
except FileNotFoundError:
    raise FileNotFoundError(f"Não encontrei o arquivo em '{csv_path}'. Verifique o caminho.")

print('Dimensão:', df.shape)
print('\nColunas:')
print(df.columns.tolist())

display(df.head(10))

# Valores ausentes
missing = df.isna().sum().sort_values(ascending=False)
missing = missing[missing > 0]
if not missing.empty:
    fig_missing = px.bar(
        missing.reset_index(), x='index', y=0,
        labels={'index': 'Coluna', '0': 'Valores ausentes'},
        title='Contagem de valores ausentes por coluna'
    )
    fig_missing.update_layout(xaxis_tickangle=-45, height=400)
    fig_missing.show()
else:
    print('\nNão há valores ausentes.')

# Seleção automática de colunas para gráficos simples
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
cat_cols = df.select_dtypes(exclude=['number']).columns.tolist()

# Histograma para a primeira coluna numérica
if numeric_cols:
    num_col = numeric_cols[0]
    fig_num = px.histogram(df, x=num_col, nbins=30, title=f'Distribuição de {num_col}')
    fig_num.update_layout(height=400)
    fig_num.show()
else:
    print('\nNão há colunas numéricas para plotar.')

# Barra para a primeira coluna categórica (top 20 categorias)
if cat_cols:
    cat_col = cat_cols[0]
    top_counts = df[cat_col].astype(str).value_counts().head(20)
    fig_cat = px.bar(
        top_counts.reset_index(), x='index', y=cat_col,
        labels={'index': cat_col, cat_col: 'Frequência'},
        title=f'Distribuição (Top 20) de {cat_col}'
    )
    fig_cat.update_layout(xaxis_tickangle=-45, height=400)
    fig_cat.show()
else:
    print('\nNão há colunas categóricas para plotar.')
```
